```python
import matplotlib.pyplot as plt


products = ["Product A", "Product B", "Product C", "Product D"]
sales = [120, 250, 180, 90]
```

```python
plt.figure()
plt.bar(products, sales)
plt.title("Sales Data - Default Colors")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
colors_high_contrast = ["red", "blue", "green", "orange"]

plt.figure()
plt.bar(products, sales, color=colors_high_contrast)
plt.title("Sales Data - High Contrast Colors")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
colors_mono = ["#003366", "#336699", "#6699CC", "#99CCFF"]

plt.figure()
plt.bar(products, sales, color=colors_mono)
plt.title("Sales Data - Monochromatic Blue Palette")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
colors_poor = ["lightgray", "silver", "gainsboro", "whitesmoke"]

plt.figure()
plt.bar(products, sales, color=colors_poor)
plt.title("Sales Data - Poor Contrast (Low Readability)")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```
